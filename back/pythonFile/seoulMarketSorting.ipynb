```python
import pandas as pd
import numpy as np

df = pd.read_csv('seoulMarket.csv', sep=',', encoding='utf-8')
```

```python
df = df.loc[(df['상권업종대분류명'] == '음식') | (df['상권업종대분류명'] == '관광/여가/오락') | ((df['상권업종대분류명'] == '소매') & ((df['상권업종중분류명'] == '종합소매점') | (df['상권업종중분류명'] == '시계/귀금속소매') | (df['상권업종중분류명'] ==  "취미/오락관련소매") | (df['상권업종중분류명'] ==  "취미/선물/팬시/기념품") ))]

df = df[['상호명', '상권업종중분류명', '상권업종소분류명', '표준산업분류명', '시군구명','행정동명', '위도', '경도']]

df = df.loc[(df['시군구명'] == '동대문구') | (df['시군구명'] == '성북구')]
```

```python
df.columns = ['name',  # 상호명
              'cate_1',  # 중분류명
              'cate_2',  # 소분류명
              'cate_3',  # 표준산업분류명
              'state',  # 시군구명
              'dong',  # 행정동명
              'lon',  # 위도
              'lat'  # 경도
              ]
```

```python
df.to_csv("sortedSeoulMarket.csv", mode='w')
```
